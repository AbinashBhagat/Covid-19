# Italy Covid-19 

### This is a dataset of Italy of Covid-19 pandemic.(1st jan 2020-3rd sep 2020)
 These contains the following columns:
 1. date
 2. new_cases
 3. new_deaths
 4. new_tests

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('italy-covid-daywise.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
covid_df = df.copy()
```

### Total number of reported deaths and cases.

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
print('Total number of cases were:',total_cases,'\n Total number of deaths were:',total_deaths)
```

```python
covid_df['date'] = pd.to_datetime(covid_df.date)
```

### Reported Cases

```python
covid_df.set_index('date', inplace = True)
plt.figure(figsize=(12,6))
covid_df.new_cases.plot()
```

Conclusion: It can be easily noticed that in the month of march and april more number of cases were reported.

### Reported Deaths

```python
plt.figure(figsize=(12,6))
covid_df.new_deaths.plot()
```

Conclusion: Deaths were at a huge rate in the month of March, April, May and June.

### Tests Conducted

```python
plt.figure(figsize=(12,6))
covid_df.new_tests.plot()
```

### Reported Cases vs Reported Deaths vs Tests Conducted


```python
plt.figure(figsize=(12,6))
sns.lineplot(data = covid_df)

```
